```python
import pandas as pd
df = pd.read_csv("wines_SPA.CSV")

df
```

```python
unique_values = df['type'].unique()

# Display the unique values
print("\nUnique values in 'type' column:")
print(unique_values)
```

```python
null_type = df['type'].isnull()

# Display rows with null values in the 'type' column
print("\nRows with null values in the 'type' column:")
print(df[null_type])
```

```python
unique_values = df['winery'].unique()

# Display the unique values
print("\nUnique values in 'winery' column:")
print(unique_values)
print("\nNumber of unique values in 'winery' column:")
print(len(unique_values))
```

```python
null_type = df['winery'].isnull()

# Display rows with null values in the 'type' column
print("\nRows with null values in the 'winery' column:")
print(df[null_type])
```

```python
unique_values = df['year'].unique()

# Display the unique values
print("\nUnique values in 'year' column:")
print(unique_values)
```

```python
null_type = df['year'].isnull()

# Display rows with null values in the 'type' column
print("\nRows with null values in the 'year' column:")
print(df[null_type])
```

```python

# Convert 'year' column to numeric, coercing errors to NaN
df['year'] = pd.to_numeric(df['year'], errors='coerce')

# Calculate the mean of the numeric 'year' column
year_mean = df['year'].mean()

# Fill NaN values in the 'year' column with the mean
df['year'] = df['year'].fillna(year_mean)

# Display the DataFrame after filling NaN values
print("\nDataFrame after filling NaN values in 'year' with the mean:")
print(df)
```

```python
null_type = df['year'].isnull()

# Display rows with null values in the 'type' column
print("\nRows with null values in the 'year' column:")
print(df[null_type])
```
